```python
import re

def eliminar_emojis(texto):
    # Expresión regular para detectar emojis
    emoji_pattern = re.compile("["
                           u"\U0001F600-\U0001F64F"  # Emoticonos
                           u"\U0001F300-\U0001F5FF"  # Símbolos y pictogramas
                           u"\U0001F680-\U0001F6FF"  # Transporte y símbolos
                           u"\U0001F700-\U0001F77F"  # Símbolos adicionales
                           u"\U0001F780-\U0001F7FF"  # Otros símbolos
                           u"\U0001F800-\U0001F8FF"  # Flechas, triángulos
                           u"\U0001F900-\U0001F9FF"  # Más pictogramas
                           u"\U0001FA00-\U0001FA6F"  # Objetos varios
                           u"\U0001FA70-\U0001FAFF"  # Adicionales
                           u"\U00002702-\U000027B0"  # Símbolos adicionales
                           u"\U000024C2-\U0001F251"
                           "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', texto)
```

```python
def extraer_hashtags(texto):
    # Limpiar emojis
    texto_sin_emojis = eliminar_emojis(texto)
    # Extraer hashtags
    lista_hashtags = re.findall(r'#(\w+)', texto_sin_emojis)
    return str(lista_hashtags)
```

```python
# Limpiar emojis y crear una columna de texto limpio
df['Texto limpio'] = df['Texto del reel'].apply(eliminar_emojis)

# Extraer hashtags y guardarlos en una nueva columna
df['Hashtags'] = df['Texto limpio'].apply(extraer_hashtags)
```

```python
df['Texto limpio']= df['Texto limpio']+'Hashtags:'+df['Hashtags']
```

```python
client=OpenAI(api_key="")
```

```python
def clasificartexto(texto):
    respuesta = client.chat.completions.create(
        messages=[
            {
                "role": "user",
                "content":f"Indicame la temática (Acontecimientos Históricos, Toponimia de Lugares,Biografías de Personajes Históricos, Arquitectura, Curiosidades Históricas) del siguiente texto solo puede ser 1 categoría:{texto}"

            }
        ],
        model="gpt-3.5-turbo",
        max_tokens=60,
        temperature=0.5,
        stop=["\n"],
    )
    categoria=(respuesta.choices[0].message.content)
    return categoria
```

```python
df['Categoria']=df['Texto limpio'].apply(clasificartexto)
```

```python
def crear_titulo_redes_sociales(texto):
    respuesta = client.chat.completions.create(
        messages=[
            {
                "role": "user",
                "content":f"Crea un título llamativo para este texto que sea adecuado \
                         para redes sociales, usando un máximo de 10 palabras y que no incluya hashtags: {texto}"

            }
        ],
        model="gpt-3.5-turbo",
        max_tokens=30,
        temperature=0.5,
        stop=["\n"],
    )
    titulo=(respuesta.choices[0].message.content)
    return titulo
```

```python
df['Titulo']=df['Texto limpio'].apply(crear_titulo_redes_sociales)
```
